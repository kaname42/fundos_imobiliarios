```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

service = Service('/usr/bin/chromedriver')
service.start()
driver = webdriver.Remote(service.service_url)


driver.get("https://www.fundsexplorer.com.br/ranking")
time.sleep(3)
df = pd.read_html(driver.page_source)[0]
soup = BeautifulSoup(driver.page_source,"html.parser")
driver.quit()
```

```python
columns = ['Fundos', 'Setor', 'Dividend Yield', 'P/VP', 'Liquidez Diária (R$)', 'Patrimônio Líquido', 'Vacância Física']
```

```python
df_new = df[columns].fillna(0)
```

```python
df_new['P/VP'] = df_new['P/VP'].str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)/100
```

```python
df_new['Dividend Yield'] = df_new['Dividend Yield'].str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)/100
```

```python
df_new['Liquidez Diária (R$)'] = df_new['Liquidez Diária (R$)'].str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)/100
```

```python
df_new['Patrimônio Líquido'] = df_new['Patrimônio Líquido'].str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)/100
```
